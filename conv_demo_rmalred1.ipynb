```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Flatten, Dense, ReLU
import tensorflow as tf
import numpy as np

#model = Sequential([
#    Input(shape=(3, 3, 1)),
#    Conv2D(1, kernel_size=(2, 2), activation='relu'),
#    Flatten(),
#    Dense(1, activation='sigmoid')
#])


# Input layer
input_layer = Input(shape=(3, 3, 1))

# Convolutional layer
custom_kernel_initializer = tf.constant_initializer(-0.5)
custom_bias_initializer = tf.constant_initializer(1.5)
conv_layer = Conv2D(1, kernel_size=(2, 2), strides=(1, 1),
                    kernel_initializer=custom_kernel_initializer,
                    bias_initializer=custom_bias_initializer)(input_layer)
conv_activation = ReLU()(conv_layer)

# Flatten layer
flatten_layer = Flatten()(conv_activation)

# Dense (fully connected) layer
dense_layer = Dense(1, activation='sigmoid',
                    #kernel_initializer='ones',
                    #bias_initializer='zeros'
                   )(flatten_layer)

model = Model(inputs=input_layer, outputs=dense_layer)

# Define a new model that outputs activations from each layer
activation_model = Model(inputs=input_layer, outputs=[conv_layer, conv_activation, flatten_layer, dense_layer])

# Input image
input_image = np.array([[[1], [1], [1]],
                        [[1], [1], [1]],
                        [[0], [0], [0]]],

                       #add more images like the 3x3 image above

                       ).reshape(1, 3, 3, 1)

# Get activations
activations = activation_model.predict(input_image)

# Print the activations for each layer
for i, activation in enumerate(activations):
    print(f"Activation Layer {i}: \n{activation}\n")

# Compile the original model
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(input_image, np.array([0]), epochs=100, verbose=1)

# Get updated output (activation of the top layer)
activation_model = Model(inputs=input_layer, outputs=[conv_layer, conv_activation, flatten_layer, dense_layer])
activations = activation_model.predict(input_image)
print(f"{activations[-1]=}")
```
